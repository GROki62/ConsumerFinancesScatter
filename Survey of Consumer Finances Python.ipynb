```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#importing necessary modules and defining them

df = pd.read_csv('sub-data2.csv') #reading and defining dataset

df.head() #display head to check if dataset is working

x = 'KIDS' # defining number of kids as x axis 
y = 'FOODAWAY' #defining amount of money spent on restaurants as y axis

plt.scatter(x= df[x], y = df[y], s = 1) #making scatter plot using data from columns of dataset
plt.title('Money Spent on Restaurants by Number of Kids') #defining title of graph
plt.xlabel('Number of Kids') #naming title of x axis
plt.ylabel('Money Spent on Restaurants (in dollars) ') #naming title of y axis of graph
```
